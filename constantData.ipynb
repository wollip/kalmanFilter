```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
data = np.random.rand(100)
```

```python
print( data.mean(), data.std())
```

# setting up initial values

```python
estimateError = np.array([[0.1]])
dataError = np.array([[0.01]])
estimate = data[0]
```

# Calculate kalman gain using estimate and data errors

a large gain means that the measurements are reliable.

a large gain comes from a small dataError look at cell below

```python
gain = estimateError / ( estimateError + dataError)
```

```python
gain
```

# Calculate Estimate

```python
estimate = estimate + gain*( data[1] - estimate)
```

# Calculate new estimate error

```python
estimateError = dataError*estimateError / ( dataError + estimateError)
```

# Combining it together

```python
# kalman filter variables
estimateError = np.array([[0.3]])
dataError = np.array([[0.25]])
estimate = data[0]
# saving estimates to plot over time
pastEstimate = []
pastEstimate.append( estimate )
pastEstimateError = []
pastEstimateError.append(estimateError)
# loop through the data
for d in data[1:]:
    gain = estimateError/( estimateError + dataError)
    estimate = estimate + gain*( d - estimate)
    pastEstimate.append( estimate )
    estimateError = dataError* estimateError / ( dataError + estimateError)
    pastEstimateError.append(estimateError)
```

```python
plt.plot( data )
plt.plot( pastEstimate )
```
